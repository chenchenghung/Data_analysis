## 4.1銷售數據

```python
#年 月 下單日期 銷售額 銷量 利潤
sales_data = data[['Order_Date', 'Sales', 'Quantity', 'Profit', 'year', 'month']]
sales_data.sample(5)
```

```python
sales_by_year=data.groupby(['year','month']).sum()
sales_by_year
```

```python
sales_2011=sales_by_year.loc[(2011,slice(None)),:].reset_index()
sales_2012=sales_by_year.loc[(2012,slice(None)),:].reset_index()
sales_2013=sales_by_year.loc[(2013,slice(None)),:].reset_index()
sales_2014=sales_by_year.loc[(2014,slice(None)),:].reset_index()
sales_2014
```

#### 4.1.1銷售額分析

```python
sales=pd.concat([sales_2011['Sales'],sales_2012['Sales'],sales_2013['Sales'],sales_2014['Sales']],axis=1)
sales.columns=['Sales_2011','Sales_2012','Sales_2013','Sales_2014']
sales.index=['Jau','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales
```

```python
sales.sum()
```

```python

# 計算成長率。年度銷售金額
# 1. 計算年度銷售總額
sales_sum = sales.sum()
sales_sum.plot(kind='bar', alpha=0.5)
plt.grid()

# 2. 成長率 
rise_12=sales_sum[1]/sales_sum[0]-1
rise_13=sales_sum[2]/sales_sum[1]-1
rise_14=sales_sum[3]/sales_sum[2]-1
rise_rate=[0,rise_12,rise_13,rise_14]


sales_sum = pd.DataFrame({'Sales_sum':sales_sum})
sales_sum['rise_rate'] = rise_rate
sales_sum
```

上表可以得知，以年度為單位做比較的話，整體的銷售在三年中已經成長接近100%。表示公司的銷售策略是有效的，再來就是依照月份來觀察

```python
sales.style.background_gradient()
```

```python
sales.plot.area(stacked=False)
```

##### 從結果來看，只購買一次的用戶只有10個，大部分用戶都會購買多次，說明回頭率很高，也驗證了上面關於該企業老客戶的維繫不錯的結論。

```python

rfm = data.pivot_table(index='Customer_ID',
                    values = ["Quantity","Sales","Order_Date"],
              aggfunc={"Quantity":"sum","Sales":"sum","Order_Date":"max"})
# 所有用户最大的交易日期为标准，求每笔交易的时间间隔即为R
# R- 距離最終日前的上一筆訂單隔了多久
rfm['R'] = (rfm.Order_Date.max() - rfm.Order_Date)/np.timedelta64(1,'D')
# 每個客戶的總銷量F，總銷售額為M
rfm.rename(columns={'Quantity':'F','Sales':'M'},inplace = True)
rfm.head()
```

```python

# 對客戶價值進行標示, 分成8個級別

# 超過平均值為1，否則為0
rfm[['R','F','M']].apply(lambda x:x-x.mean())

def rfm_func(x):
    level =x.apply(lambda x:'1'if x>0 else '0')
    level =level.R +level.F +level.M
    d = {
        "111":"重要價值客戶",
        "011":"重要維持客戶",
        "101":"重要挽留客戶",
        "001":"重要發展客戶",
        "110":"一般價值客戶",
        "010":"一般維持客戶",
        "100":"一般挽留客戶",
        "000":"一般發展客戶"
    }
    result = d[level]
    return result

rfm['label']= rfm[['R','F','M']].apply(lambda x:x-x.mean()).apply(rfm_func,axis =1)
rfm.head()
```

```python
#可視化 一般客戶與重要客戶
rfm.loc[rfm.label=='重要價值客戶','color']='g'
rfm.loc[~(rfm.label=='重要價值客戶'),'color']='r'
rfm.plot.scatter('F','R',c= rfm.color)
```

#### 4.3.4新用戶, 活躍用戶, 不活躍用戶, 回歸用戶

```python

# 設置Customer_ID為索引，month為列名，統計每個月的購買次數。
pivoted_counts = data.pivot_table(index= 'Customer_ID',
                               columns= 'month',
                               values= 'Order_Date',
                               aggfunc= 'count').fillna(0)
# 若大於一次則全部填1
data_purchase = pivoted_counts.applymap(lambda x:1 if x>0 else 0)
data_purchase.head()
```

##### 將用戶按照每一個月份分成:
##### unreg:觀望用戶（前兩月沒買，第三個月才第一次買,則用戶前兩個月為觀望用戶)
##### unactive:首月購買後，後序月份沒有購買則在沒有購買的月份中該用戶的為非活躍用戶
##### new:當前月就進行首次購買的用戶在當前月為新用戶
##### active:連續月份購買的用戶在這些月中為活躍用戶
##### return:購買之後間隔n月再次購買的第一個月份為該月份的回頭客

```python
#使用狀態標籤把上圖的表轉換
def active_status(data):
    status = [] #某用戶的狀態表
    for i in range(48):
        #若當月消費為0
        if data[i] == 0:
            #如果不是第一個月
            if len(status)>0:
                #上個月未註冊
                if status[i-1] == "unreg": 
                 # 依然標記未註冊
                    status.append("unreg")
                else:
                 # 其餘的直接標記unactive
                    status.append("unactive")
            else:
                status.append("unreg")
            
        # 若當月消費>0
        else:
            #如果不是第一個月
            if len(status)>0:
                if status[i-1] == "unactive":
                 # 回頭客戶
                    status.append("return")
                #當月消費但上個月為未註冊
                elif status[i-1] == "unreg":
                    status.append("new")
                else:
                    status.append("active")
            #則當月為第一個月消費
            else:
                status.append("new")
    return pd.Series(status)  

purchase_stats = data_purchase.apply(active_status,axis =1)
purchase_stats.head()
```

```python

# 用NaN代替 "unreg"，統計每月的狀態客戶人數。
purchase_stats_ct = purchase_stats.replace('unreg',np.NaN).apply(lambda x:pd.value_counts(x))
# 用0填充NaN
purchase_stats_ct.fillna(0).T.plot.area()
```

从以上结果可以发现活跃客户、新客户和回归客户，每年呈一定的规律起伏，这可能和年终年末大促有关，需要更多数据进行佐证；同时可以发现新客数量每年均在减少，说明该商家新客获取率较低，如果能在新客户获取上取得突破，会给商家带来很大的增长空间。

#### 4.3.5 複構率與回購率指標

```python
# 複購率: 用戶在當月下單超過一次
purchase_r = pivoted_counts.applymap(lambda x :1 if x>1 else np.NaN if x==0 else 0)
(purchase_r.sum()/purchase_r.count()).plot(figsize=(10,4))
```

```python
# 回購率指標: 當月買過且下個月也買
def purchase_back(data):
    status=[]
    for i in range(47):
        if data[i] ==1:
            if data[i+1] == 1:
                status.append(1)
            if data[i+1] == 0:
                status.append(0)
        else:
            status.append(np.NaN)
    status.append(np.NaN)
    return status

purchase_b = data_purchase.apply(purchase_back,axis =1,result_type='expand')
(purchase_b.sum()/purchase_b.count()).plot(figsize=(10,4))
```
